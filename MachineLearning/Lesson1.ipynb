# üß† Concept Recap ‚Äî Machine Learning

## üîπ What is Machine Learning?
Machine Learning is a branch of **Artificial Intelligence** where computers
**learn patterns from data** and make predictions **without being explicitly programmed**.

---

## üîπ Traditional Programming vs ML

| Traditional Programming | Machine Learning |
|-------------------------|------------------|
| Rules + Data ‚Üí Output   | Data + Output ‚Üí Rules |
| Programmer writes logic| Model learns logic |

---

## üîπ Types of Machine Learning
- **Supervised Learning** ‚Üí Labeled data (e.g., spam detection)
- **Unsupervised Learning** ‚Üí Unlabeled data (e.g., clustering)
- **Reinforcement Learning** ‚Üí Reward-based learning (e.g., games, robots)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Simple dataset: study hours vs marks
hours = np.array([1,2,3,4,5,6]).reshape(-1,1)
marks = np.array([35,40,50,60,65,75])

# Train model
model = LinearRegression()
model.fit(hours, marks)

# Predict for new student
new_hours = [[7]]
predicted_marks = model.predict(new_hours)

print("Predicted marks for 7 hours study:", predicted_marks[0])
```

```python
plt.scatter(hours, marks)
plt.plot(hours, model.predict(hours))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Machine Learning Prediction Example")
plt.show()
```

```python
# Try changing hours and see prediction
h = int(input("Enter study hours: "))
print("Predicted Marks:", model.predict([[h]])[0])
```
