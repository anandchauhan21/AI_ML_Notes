# üß† Concept Recap ‚Äî Types of Machine Learning

## 1Ô∏è‚É£ Supervised Learning
- Uses **labeled data** (input + output known)
- Goal: Learn mapping from X ‚Üí Y  
- Examples: House price prediction, spam detection

## 2Ô∏è‚É£ Unsupervised Learning
- Uses **unlabeled data**
- Finds hidden patterns or groups
- Examples: Customer segmentation, topic modeling

## 3Ô∏è‚É£ Reinforcement Learning
- Learns by **trial and error**
- Uses **rewards & penalties**
- Examples: Game playing AI, robot navigation


# üîç Comparison of ML Types

| Feature | Supervised | Unsupervised | Reinforcement |
|--------|------------|--------------|---------------|
| Data | Labeled | Unlabeled | No fixed dataset |
| Output | Known | Unknown | Reward signal |
| Goal | Predict output | Find patterns | Maximize reward |
| Example | Email spam filter | Customer groups | Self-driving car |


```python
# Supervised Learning
from sklearn.linear_model import LinearRegression
import numpy as np

# Study hours vs marks
X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([35,45,55,65,75])

model = LinearRegression()
model.fit(X, y)

print("Predicted marks for 6 hours:", model.predict([[6]])[0])
```

```python
# Unsupervised Learning
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Simple dataset
X = np.array([[1,2],[2,1],[3,3],[8,8],[9,9],[8,9]])

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.title("Unsupervised Learning: Clustering")
plt.show()
```

```python
# Reinforcement Learning
import random

actions = ["Left", "Right"]
reward = 0

for step in range(5):
    action = random.choice(actions)
    if action == "Right":
        reward += 10
    else:
        reward -= 5
    print(f"Step {step+1}: Action = {action}, Reward = {reward}")
```

```python
# Try changing the input
h = int(input("Enter study hours: "))
print("Predicted Marks:", model.predict([[h]])[0])
```
