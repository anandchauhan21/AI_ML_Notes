Linear Algebra is the **mathematical foundation** of Machine Learning.

---

## ðŸ”¹ Why Linear Algebra?
- Data is stored as **vectors & matrices**
- Model parameters are **matrices**
- Predictions involve **matrix multiplication**

---

## ðŸ”¹ Key Concepts
- Scalars
- Vectors
- Matrices
- Matrix operations
- Dot product

```python
import numpy as np

# Scalar
a = 5

# Vector
v = np.array([1, 2, 3])

# Matrix
M = np.array([[1, 2], [3, 4]])

print("Scalar:", a)
print("Vector:", v)
print("Matrix:\n", M)
```

```python
v1 = np.array([2, 4])
v2 = np.array([1, 3])

print("Addition:", v1 + v2)
print("Subtraction:", v1 - v2)
print("Scalar Multiplication:", 3 * v1)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A + B:\n", A + B)
print("A - B:\n", A - B)
print("A Ã— B:\n", A @ B)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("Dot Product:", np.dot(x, y))
```

```python
C = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", C)
print("Transpose:\n", C.T)
```

```python
# Simple linear model y = wx + b
w = np.array([2, 3])
x = np.array([1, 2])
b = 1

y = np.dot(w, x) + b
print("Model Output:", y)
```

```python
# Try changing vector values
x = np.array([3, 4])
w = np.array([1, 2])
b = 2

print("Prediction:", np.dot(w, x) + b)
```
